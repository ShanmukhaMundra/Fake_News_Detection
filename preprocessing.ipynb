```python
data = pd.concat([fake_data, true_data], ignore_index=True)
```

```python
data = data.drop(['title', 'subject', 'date'], axis=1)
```

```python
import re
import string
```

```python
def tokenize_text(text):
    text = text.lower()
    text = re.sub(r"\\W", " ", text)
    text = re.sub(r"https?://\S+|www\.\S+", " ", text)
    text = re.sub(r"<.*?>+", " ", text)
    text = re.sub(r"[%s]" % re.escape(string.punctuation), " ", text)
    text = re.sub(r"\n", " ", text)
    text = re.sub(r"\[.*?\]", " ", text)
    return text
```

```python
data['text'] = data['text'].apply(tokenize_text)
```

```python
data.head(10)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
```

```python
x = data['text']
y = data['class']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
vector = TfidfVectorizer()
x_train = vector.fit_transform(x_train)
x_test = vector.transform(x_test)
```

```python
from sklearn.linear_model import LinearRegression
linear_regression = LinearRegression()
linear_regression.fit(x_train, y_train)
```

```python
predictions_lr = linear_regression.predict(x_test)
accuracy_lr = linear_regression.score(x_test, y_test)
accuracy_lr = round(accuracy_lr, 2)
```

```python
accuracy_lr
```

```python
from sklearn.linear_model import LogisticRegression
logistic_regression = LogisticRegression()
logistic_regression.fit(x_train, y_train)
```

```python
predictions_lgr = logistic_regression.predict(x_test)
accuracy_lgr = logistic_regression.score(x_test, y_test)
accuracy_lgr = round(accuracy_lgr, 2)
```

```python
accuracy_lgr
```

```python
from sklearn.tree import DecisionTreeClassifier
d_tree = DecisionTreeClassifier()
d_tree.fit(x_train, y_train)
```

```python
predictions_dtree = d_tree.predict(x_test)
accuracy_dtree = d_tree.score(x_test, y_test)
accuracy_dtree = round(accuracy_dtree, 2)
```

```python
accuracy_dtree
```

```python
from sklearn.ensemble import GradientBoostingClassifier, RandomForestClassifier

g_boosting = GradientBoostingClassifier()
g_boosting.fit(x_train, y_train)
```

```python
predictions_gboost = g_boosting.predict(x_test)
accuracy_gboost = g_boosting.score(x_test, y_test)
accuracy_gboost = round(accuracy_gboost, 2)
```

```python
accuracy_gboost
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
rf.fit(x_train, y_train)
```

```python
predictions_rf = rf.predict(x_test)
accuracy_rf = rf.score(x_test, y_test)
accuracy_rf = round(accuracy_rf, 2)
```
